```python
def sort_arr(score, track):
    score_copy = score.copy()
    track_copy = track.copy()
    for j in range(len(score_copy)):
        for i in range(j+1,len(score_copy)):
            if score_copy[j]>score_copy[i]:
                temp = score_copy[j]
                score_copy[j]=score_copy[i]
                score_copy[i]=temp
                temp1=track_copy[j]
                track_copy[j]=track_copy[i]
                track_copy[i]=temp1
    return score_copy,track_copy
def assign_val(track1, track2):
    val = 1
    val_list = [0]*6
    counter = 3
    index = len(track1)-1
    while index>-1:
        for i in range(0, len(track2)):
            if track1[index]==track2[i]:
                if counter>0:
                    val = 1
                    val_list[i]=val
                    counter = counter - 1
                else:
                    val = 0
                    val_list[i]=val
        index = index -1
    return val_list
```
