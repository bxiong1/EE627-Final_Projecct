```python
import csv
def addition(a1,start,end):
    result =0
    for i in range(start,end):
        result = result+float(a1[i])
    return result
def averages(a1,start,end):
    result =0
    count = 0
    for i in range(start,end):
        result = result+float(a1[i])
        if float(a1[i])>0:
            count = count+1
    if count == 0:
        count = 1
    else:
        count = count
    return result/count
def sort_arr(score, track):
    score_copy = score.copy()
    track_copy = track.copy()
    for j in range(len(score_copy)):
        for i in range(j+1,len(score_copy)):
            if score_copy[j]>score_copy[i]:
                temp = score_copy[j]
                score_copy[j]=score_copy[i]
                score_copy[i]=temp
                temp1=track_copy[j]
                track_copy[j]=track_copy[i]
                track_copy[i]=temp1
    return score_copy,track_copy
def assign_val(track1, track2):
    val = 1
    val_list = [0]*6
    counter = 3
    index = len(track1)-1
    while index>-1:
        for i in range(0, len(track2)):
            if track1[index]==track2[i]:
                if counter>0:
                    val = 1
                    val_list[i]=val
                    counter = counter - 1
                else:
                    val = 0
                    val_list[i]=val
        index = index -1
    return val_list
def assign_val_1(track1, track2, val_lists):
    val = 1
    val_list = [0]*6
    index = len(track1)-1
    while index>-1:
        for i in range(0, len(track2)):
            if track1[index]==track2[i]:
                if float(val_lists[i])>0:
                    val = 1
                    val_list[i]=val
                else:
                    val = 0
                    val_list[i]=val
        index = index -1
    return val_list
def write_csv(userID,trackID,val_list,num):
    name = list()
    for i in range(0,len(trackID)):
        name.append(str(userID[i])+'_'+str(trackID[i]))
    with open('result'+str(num)+'.csv',"wt",newline="") as f:
        writer = csv.writer(f,delimiter=",")
        writer.writerow(["TrackID","Predictor"])
        for i in range(0, len(val_list)):
            writer.writerow([name[i],val_list[i]])
def my_sum(l):
    sum=0
    count = 0
    for i in l:
        if int(i) != 0:
            count+=1
        sum=sum+int(i)
    if count != 0:
        return sum/count
    else:
        return sum
```
